

**Dataset Overview**:
    - The dataset contains 1000 entries and 21 columns.
    - There are no missing values in any of the columns.

1. **Categorical Features**:
    - There are 13 categorical features: `status_of_existing_checking_account`, `credit_history`, `purpose`, `savings_account_bonds`, `present_employment_since`, `personal_status_and_sex`, `other_debtors_guarantors`, `property`, `other_installment_plans`, `housing`, `job`, `telephone`, and `foreign_worker`.
    - These features will need to be encoded before training the model.

2. **Numerical Features**:
    - There are 8 numerical features: `duration_in_month`, `credit_amount`, `installment_rate_in_percentage_of_disposable_income`, `present_residence_since`, `age_in_years`, `number_of_existing_credits_at_this_bank`, `number_of_people_being_liable_to_provide_maintenance_for`, and `creditability`.
    - These features can be used directly for training, but may benefit from normalization or scaling.



**Potential Data Imbalance**:
    - The target variable `creditability` should be checked for class imbalance, which may require techniques such as resampling or using appropriate evaluation metrics.


```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Analyze categorical features
categorical_features = df.select_dtypes(include=['object']).columns
for feature in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=feature, hue='creditability')
    plt.title(f'Distribution of {feature}')
    plt.xticks(rotation=90)
    plt.legend(title='Creditability')
    plt.show()
```

```python
# Analyze numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
for feature in numerical_features:
    plt.figure(figsize=(10, 5))
    histplot = sns.histplot(data=df, x=feature, hue='creditability', kde=True, multiple='stack')
    plt.title(f'Distribution of {feature}')
    plt.legend(title='Creditability', labels=['Bad', 'Good'])
    plt.show()
```

```python
# Correlation analysis
plt.figure(figsize=(12, 8))
numerical_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numerical_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
